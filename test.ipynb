```python
#财报测试
import requests
import json

# 配置信息
API_URL = "https://www.baicai.chat/v1/chat/completions"
API_KEY = "sk-KmWptQm8Lg6BYcWecVdq2OJ76NfmLIXRTC07K0YhTBEm82SF"
MODEL = "[M]gemini-3-pro-preview" # 选一个性能最强的

def generate_financial_summary():
    # 1. 读取改进版英文 Prompt
    with open('Prompt.txt', 'r', encoding='utf-8') as f:
        system_prompt = f.read()

    # 2. 读取财报 TXT 内容 (截取前 150,000 字符以覆盖核心财务数据)
    with open('2025-Q4-10-Q.txt', 'r', encoding='utf-8') as f:
        report_content = f.read()

    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }

    payload = {
        "model": MODEL,
        "messages": [
            {"role": "user", "content": f"{system_prompt}\n\nReport Content:\n{report_content}"}
        ]
    }



    print(f"正在使用 {MODEL} 处理财报摘要，请稍候...")
    
    try:
        response = requests.post(API_URL, headers=headers, json=payload, timeout=300)
        
        if response.status_code == 200:
            result = response.json()
            summary = result["choices"][0]["message"]["content"]
            
            # 将摘要保存到本地文件
            with open('ADBE_Q2_Summary_EN.md', 'w', encoding='utf-8') as f:
                f.write(summary)
            
            print("\n--- 摘要生成成功 (已保存至 ADBE_Q2_Summary_EN.md) ---")
            print(summary)
        else:
            print(f"请求失败: {response.status_code}, {response.text}")
            
    except Exception as e:
        print(f"发生错误: {e}")

if __name__ == "__main__":
    generate_financial_summary()
```
